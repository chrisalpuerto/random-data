```python
import pandas as pd
```

```python
# Data from sesamesim.csv using python
file_path = "public/"
df = pd.read_csv(f"{file_path}sesamesim.csv")
row, col = df.shape
print(f"Rows: {row}, Cols: {col}")
table = df.groupby("site")["prenumb"].describe()
print("Data Per Site:")
print(f"{table}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# overall histogram
plt.figure(figsize=(5,2))
sns.histplot(df["prenumb"], kde=True)
plt.title("Distribution of pre-test number scores")
plt.show()

# boxplot by site
plt.figure(figsize=(5,2))
sns.boxplot(x="site", y="prenumb", data=df)
plt.title("Pre-Test scores by site")
plt.show()
```

```python
# Scatterplot with trendline
plt.figure(figsize=(5,3))
sns.regplot(x="prenumb", y="postnumb", data=df)
plt.title("Relationship Between Pre- and Post-Test Number scores")
plt.show()
r = df['prenumb'].corr(df['postnumb'])
print("Correlation: ", round(r,3))
```

```python
def find_sse(arr, _mean):
    res = 0
    for num in arr:
        res += (num - _mean) ** 2
    return res
fert_a = [22,25,24,23,26]
fert_b = [18,21,19,20,22]
fert_c = [27,30,28,29,26]

mean_a = sum(fert_a) // 5
mean_b = sum(fert_b) // 5
mean_c = sum(fert_c) // 5

fert_a_sse = find_sse(fert_a, mean_a)
fert_b_sse = find_sse(fert_b, mean_b)
fert_c_sse = find_sse(fert_c, mean_c)

print(f"SSE fert a: {fert_a_sse}")
print(f"SSE fert b: {fert_b_sse}")
print(f"SSE fert c: {fert_c_sse}")
```
